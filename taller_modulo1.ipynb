### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

vida = [72, 75, 80, 65, 90, 85, 78, 60, 74, 88]

# Promedio
suma = 0
for v in vida:
    suma += v

promedio = suma / len(vida)
print("Promedio:", promedio)

# Máximo
maximo = vida[0]
for v in vida:
    if v > maximo:
        maximo = v

print("Máximo:", maximo)

# Mínimo
minimo = vida[0]
for v in vida:
    if v < minimo:
        minimo = v

print("Mínimo:", minimo)

````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí

def clasificar_esperanza_de_vida(años):
    if años >= 75:
        return "Alta"
    elif años >= 60:
        return "Media"
    else:
        return "Baja"
    
# Prueba de la función

print(clasificar_esperanza_de_vida(55))  # Baja
print(clasificar_esperanza_de_vida(65))  # Media
print(clasificar_esperanza_de_vida(80))  # Alta 
print(clasificar_esperanza_de_vida(75))  # Alta
print(clasificar_esperanza_de_vida(59.9))  # Baja
````

**Responde:**

**¿Qué archivos debería incluir un `.gitignore` en un proyecto de Data Science? Menciona al menos 3 y explica por qué.**

> Tu respuesta:
**-Archivos temporales de cache:** Son archivos generadosautomaticamente por python o jupiter los cuales no aportan valor a el codigo fuente y puden llegar a generar conflictos en git.
__pycache__/
*.pyc
.ipynb_checkpoints/

.DS_Store
Thumbs.db
